```python
def random_graph (n):
    import random
    import networkx as nx
    import math
    c = random.uniform(0,2)
    if c>1:
        p=random.uniform(c*math.log(n)/n,1)
    else:
        p=random.uniform(0,c*math.log(n)/n)
    rg=nx.gnp_random_graph(n,p)
    return (nx.draw(rg,add_labels=True),c)
```

```python
random_graph(20)
```
